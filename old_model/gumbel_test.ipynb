```python
from scipy.stats import gumbel_r
import matplotlib.pyplot as plt
```

```python
env_knowledge_chance = gumbel_r.rvs(loc = 0, scale = 0.5, size = 10000)

print(env_knowledge_chance)

plt.hist(env_knowledge_chance, bins = 30, density = True)
plt.axvline(0.1, color = 'g')
# plt.axvline(1, color = 'g')
```
